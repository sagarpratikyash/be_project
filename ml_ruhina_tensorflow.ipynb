```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
batchSize = 100
epochs = 1
n_classes = 6
```

```python
def one_hot(y):
    r = np.zeros(n_classes)
    if y == 0:
        r[0] = 1
    elif y == 1:
        r[1] = 1
    elif y == 2:
        r[2] = 1
    elif y == 3:
        r[3] = 1
    elif y == 4:
        r[4] = 1
    else:
        r[5] = 1
    
    return list(r)
        
    
```

```python
data2 = pd.DataFrame()
data2['temporary_symptoms'] =train['diagnosis'].apply(lambda x : one_hot(x))
```

```python
train[["y"+str(i) for i in range(6)]] = pd.DataFrame(data2.temporary_symptoms.values.tolist(), index = data2.index)
```

```python
from sklearn.model_selection import train_test_split
train = train.dropna()
train, test = train_test_split(train, test_size = 0.1)
```

```python
X_train=train[['age','sex','s0','s1','s2','s3','s4','s5','s6']]
Y_train=train[['y0','y1','y2','y3','y4','y5']]
X_test=test[['age','sex','s0','s1','s2','s3','s4','s5','s6']]
Y_test=test[['y0','y1','y2','y3','y4','y5']]
```

```python
batchSize = 100
epochs = 1
def get_train_batch(i):
    i = i * batchSize
    x = X_train[i:i+batchSize]
    y = Y_train[i:i+batchSize]

    return x, y

def get_test_batch(i):
    i = i * batchSize
    x = X_test[i:i+batchSize]
    y = Y_test[i:i+batchSize]

    return x, y
    
```

```python
#import tensorflow as tf
#from tensorflow.examples.tutorials.mnist import input_data
#mnist = input_data.read_data_sets("/tmp/data/", one_hot = True)

#The nodes
n_nodes_hl1 = 500
n_nodes_hl2 = 500
n_nodes_hl3 = 500

#output nodes


#Batch
batch_size = 100

#The place holder to hold the values of new x and y at different times.
x = tf.placeholder('float', [100, 9])
y = tf.placeholder('float', [100, n_classes])

def neural_network_model(data):
    hidden_1_layer = {'weights':tf.Variable(tf.random_normal([9, n_nodes_hl1])),
                      'biases':tf.Variable(tf.random_normal([n_nodes_hl1]))}

    hidden_2_layer = {'weights':tf.Variable(tf.random_normal([n_nodes_hl1, n_nodes_hl2])),
                      'biases':tf.Variable(tf.random_normal([n_nodes_hl2]))}

    hidden_3_layer = {'weights':tf.Variable(tf.random_normal([n_nodes_hl2, n_nodes_hl3])),
                      'biases':tf.Variable(tf.random_normal([n_nodes_hl3]))}

    output_layer = {'weights':tf.Variable(tf.random_normal([n_nodes_hl3, n_classes])),
                    'biases':tf.Variable(tf.random_normal([n_classes])),}


    l1 = tf.add(tf.matmul(data,hidden_1_layer['weights']), hidden_1_layer['biases'])
    l1 = tf.nn.relu(l1)

    l2 = tf.add(tf.matmul(l1,hidden_2_layer['weights']), hidden_2_layer['biases'])
    l2 = tf.nn.relu(l2)

    l3 = tf.add(tf.matmul(l2,hidden_3_layer['weights']), hidden_3_layer['biases'])
    l3 = tf.nn.relu(l3)

    output = tf.matmul(l3,output_layer['weights']) + output_layer['biases']

    return output

def train_neural_network(x):
    prediction = neural_network_model(x)#the forward feed
   
    cost = tf.reduce_mean( tf.nn.softmax_cross_entropy_with_logits(logits=prediction, labels=y) )#The cost
    optimizer = tf.train.AdamOptimizer().minimize(cost)#The optimizer
    
    hm_epochs = 2 # Forward+Backprop
    with tf.Session() as sess:
       
        sess.run(tf.global_variables_initializer())#Initialize Variables

        for epoch in range(hm_epochs):
            epoch_loss = 0
            for i in range(int(len(X_train)/batch_size)):
               # print(i)
                epoch_x, epoch_y = get_train_batch(i)
                _,c,p= sess.run([optimizer, cost,prediction], feed_dict={x: epoch_x, y: epoch_y})
                epoch_loss += c
               # print(p)
            print('Epoch', epoch, 'completed out of',hm_epochs,'loss:',epoch_loss)
        #Evaluation
        correct = tf.equal(tf.argmax(prediction, 1), tf.argmax(y, 1))
        
        accuracy = tf.reduce_mean(tf.cast(correct, 'float'))
        test_x, test_y = get_test_batch(1)
        print('Accuracy:',sess.run([accuracy,tf.argmax(prediction, 1),tf.argmax(y, 1)],feed_dict={x: test_x, y: test_y}))

train_neural_network(x)

#Flow of the graph is from the session start to whatever it requires.
```
